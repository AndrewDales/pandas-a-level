```python
import pandas as pd
import matplotlib.pyplot as plt

nac = pd.read_csv("national_alevel_counts.csv")
```

```python
nac_all = nac[(nac["characteristic_type"] == "All students") &
              (nac["characteristic_value"] == "All students") &
              (nac["characteristic_gender"] == "All genders")
]
nac_all.sample(10)
```

```python
df = nac_all.drop(columns="total_students").groupby(by="time_period").mean()
df.plot(kind='bar', figsize=(10, 8))
# ax = nac_all.groupby(by="time_period").plot(kind='bar', x="subject_sector_area", y="astar_grade_achieved")
```

```python
dft = df.transpose()
fig = dft.plot(kind="pie", autopct=lambda p: f'{p:.1f}%', subplots=True, figsize=(12, 8), labels=None, legend=None)
```
